```python
import pandas as pd
# Read Excel file

data = pd.read_excel("asn-data.xlsx", "GenInfo (2)")
# df
```

```python
# Create DataFrame variable and drop columns

df = pd.DataFrame(data)
columns_to_keep = ['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 15',
                   'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 21', 'Unnamed: 27',
                   'Unnamed: 28', 'Unnamed: 29', 'Unnamed: 30', 'Unnamed: 31', 'Unnamed: 32', 'Unnamed: 33',
                   'Unnamed: 34', 'Unnamed: 35', 'Unnamed: 39', 'Unnamed: 41', 'Unnamed: 42', 'Unnamed: 43',
                   'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 47', 'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50',
                   'Unnamed: 51', 'Unnamed: 52', 'Unnamed: 53', 'Unnamed: 54', ]
# size = len(columns_to_keep)
# print(size)
df.drop(df.columns.difference(columns_to_keep), 1, inplace=True)
# df
```

```python
# drop rows

df.drop(df.index[[0, 1, 2, 4, 5]], inplace=True)
df.loc[3, 'Unnamed: 42'] = 'Ethn'
# df
```

```python
# drop empty rows at the end of file

df = df.iloc[:51]
df
```

```python
df.columns = df.iloc[0]
df = df.iloc[1:].reset_index(drop=True)
# df.head()
```

```python
# save as csv

df.to_excel('asn_clean_data.xlsx', index=False)
df.to_csv('asn_clean_data.csv', index=False)
```
