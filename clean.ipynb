```python
import pandas as pd
import uuid
# Read Excel file

data = pd.read_excel("asn-data.xlsx", "GenInfo (2)")
```

```python
# Create DataFrame variable and drop columns

df = pd.DataFrame(data)
columns_to_keep = ['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 8', 
                   'Unnamed: 9', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 
                   'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 
                   'Unnamed: 25', 'Unnamed: 27', 'Unnamed: 28', 'Unnamed: 29', 'Unnamed: 30', 'Unnamed: 31', 
                   'Unnamed: 32', 'Unnamed: 33', 'Unnamed: 34', 'Unnamed: 35', 'Unnamed: 39', 'Unnamed: 41',
                   'Unnamed: 42', 'Unnamed: 43', 'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 47', 'Unnamed: 48', 
                   'Unnamed: 49', 'Unnamed: 50', 'Unnamed: 51', 'Unnamed: 52', 'Unnamed: 53', 'Unnamed: 54', ]
# size = len(columns_to_keep)
# print(size)
df.drop(df.columns.difference(columns_to_keep), 1, inplace=True)
df
```

```python
# drop rows

df.drop(df.index[[0, 1, 2, 4, 5]], inplace=True)
df.loc[3, 'Unnamed: 42'] = 'Et®hn'
# df
```

```python
# drop empty rows at the end of file

df.dropna(how='all', inplace=True)
df
```

```python
df.columns = df.iloc[0]
df = df.iloc[1:].reset_index(drop=True)
df["Import UUID"] = [uuid.uuid4() for x in range(len(df))]
df.head()
```

```python
# clean columns
df[["Caseworker First Name", "Caseworker Last Name"]] = df["Current Caseworker Name"].str.split(" ", expand = True)
df["Street Address"] = df["Street Address"] + df["Unit / Suite #"]

df.drop(['Unit / Suite #', "Current Caseworker Name", "DAYS at current placement"], axis=1, inplace = True)

salesforce_meta_data = ["OFFICIALLY PENDING START", "OFFICIALLY PENDING END", "REFERRAL END STATUS", "REFERRAL END STATUS DATE" "IF NOT ACCEPTED OR NOT APPROVED BY DCFS, PLEASE PROVIDE REASON",]
```

```python
# Rename the columns to match Salesforce
renaming_dict = {
    "FIRST NAME" : "First Name",
    "LAST NAME" : "Last Name",
    "DCFS ID#" : "DCFS ID",
    "EMAIL ADDRESS" : "Email",
    "PHONE #" : "Contact: Phone",
    "ENROLLMENT DATE": "Date Enrolled",
    "Number of Previous HS" : "Number of previous high schools",
    "Numnber of DAYS out of school" : "Number of days out of school",
    "Number of Placements" : "Total Number of Placements",
    "Phone": "Caseworker Phone Number",
    "Email": "Caseworker Email",
    "Agency": "Caseworker Agency",
    "Street Address": "Caseworker Street Address",
    "City": "Caseworker City",
    "State": "Caseworker State",
    "Zip": "Caseworker Zip Code",
    "DOB": "Birthdate",
    "Et®hn": "Ethnicity",
    "LIVING SITUATION": "Student Living Situation Type",
    "Number of DAYS out of school": "Number of days out of school",
    "SpecEd" : "Special Needs Status",
    "DUE DATE IF PREGNANT?": "Student Pregnancy Due Date",
    "# CHILDREN" : "Number of children",
    "Grade at Entry": "Grade Level",
    "SCHOOL": "Schools",
    "CURRENT STATUS": "Enrollment Status",
    "DATE REFERRED TO ASN": "Date Referral Recieved At School",
    "DATE APPROVED BY DCFS": "Date Approved By DCFS",
    "Gender": "Gender",
    "IF NOT ACCEPTED OR NOT APPROVED BY DCFS, PLEASE PROVIDE REASON": "Reason for Rejection"
}
df.rename(columns=renaming_dict, inplace=True)
df.columns
```

```python
#split for testing
df = df.sample(n=3)
```

```python
# Map columns to objects

student_contact_df = df[["Last Name", "First Name", 'Email','Contact: Phone', 'Birthdate', 'Gender', 'Ethnicity', 'Import UUID', 'DCFS ID']]
ys3_columns = [column for column in list(df.columns) if column not in list(student_contact_df.columns) or 'Import UUID'] 
ys3_program_df = df[ys3_columns]

```

```python
# Clean Data

ys3_program_df["SCHOOL EXIT DATE"] = ys3_program_df["SCHOOL EXIT DATE"].apply(lambda x: x if x != 'Current' else None)
ys3_program_df["Caseworker Zip Code"] = ys3_program_df["Caseworker Zip Code"].astype('string')
ys3_program_df["Number of days out of school"] = ys3_program_df["Number of days out of school"].apply(lambda x: x if x != 'Unk' else None)
ys3_program_df["Total Number of Placements"] = ys3_program_df["Total Number of Placements"].apply(lambda x: x if x != 'Unk' else None)
ys3_program_df.rename(columns={"Import UUID": "Contact"}, inplace=True)
ys3_program_df["Caseworker Zip Code"]
```

```python
# save as csv
student_contact_df.to_csv('contact_import_data.csv', index=False)
ys3_program_df.to_csv('ys3_program_import_data.csv', index=False)
```
